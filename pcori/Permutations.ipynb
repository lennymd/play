Using Python to find all the permutations I need to consider for screens


```python
import itertools
```

```python
categories = ["t", "n", "c", "o","q"]
```

```python
combo1 = list(itertools.combinations(categories, 1))
combo1
```

```python
combo2 = list(itertools.combinations(categories,2))
combo2
```

```python
len(combo2)
```

```python
combo3 = list(itertools.combinations(categories,3))
print(len(combo3))
combo3
```

```python
combo4 = list(itertools.combinations(categories,4))
print(len(combo4))
combo4
```
