Let's get all the possible combinations together and then clean things up

```python
all_combos_list = []
for i in range(1,6):
    all_combos_list.append(list(itertools.combinations(all_options, i)))
```

Now that they're all in a list, let's go through that list and extract the combos:

```python
all_combos = []
for each_list in all_combos_list:
    for combo in each_list:
        all_combos.append(combo)
        
all_combos
```

Now that we have the combos, let's attempt to clean them. We'll check in each combo how many of each category or menu item we have. If there's more than one, we will not keep the combination. We don't care about order right now.

```python
import numpy as np
```

```python
good_combinations = []
for combo in all_combos:
    count_dict = {}
    for item in combo:
        category = item[0]
        if category not in count_dict:
            count_dict[category] = 1
        else:
            count_dict[category] +=1
            
    checkCount = [count_dict[key]<2 for key in count_dict]
    keep = np.prod(checkCount)
    if keep == 1:
        good_combinations.append(list(combo))
good_combinations[100:130]
```

How many good combinations are there, without accounting for order?

```python
len(good_combinations)
```

Now that we have the good ones, let's get the unique ones, taking order into account.

```python
unique_screens = []
sorted_combos = []
test = [('T0', 'O13'),('T0', 'O14'),('O13', 'T0')]
for combo in good_combinations:
    i = good_combinations.index(combo)
    ordered = sorted(combo)
    if ordered in sorted_combos:
        print(i,combo, ordered)
        pass
    else:
        sorted_combos.append(ordered)
        unique_screens.append(combo)

print(len(sorted_combos),len(unique_screens))
```
